# Mushroom Project:

# Data Set Information:

This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy.

Gill size of mushroom

Gill color of mushroom

A slender upwright object or supporting or connecting part

14.Stalk-Surface-Below-Ring

Color of stalk above ring

16.Stalk-color-below-ring

18.Veil-color

Temporary structure of tissue found on the fruiting bodies of some basidiomycete fungi, typically agarics.

19.Ring-number

A circle of mushrooms in a grassy area.

20.Ring-type

Fairy rings are mushrooms that appear in a circular formation, usually in forests or grassy areas.

21.Spore-print-color

The spore print is the powdery deposit obtained by allowing spores of a fungal fruit body to fall onto a surface underneath.

22.Population

The population of Mushrooms in certain region

23.Habitat

Types of Mushrooms grown in different region.

# Since the outcome has only two variable we will use binary classification model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import accuracy_score
```

# Splitting the dataframe:

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(trainx,trainy,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print('At random state',i,'The model performs very well')
        print('At random state',i)
        print('Training r2_score is:-',accuracy_score(y_train,pred_train)*100)
        print('Testing r2_score is:-',accuracy_score(y_test,pred_test)*100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(trainx,trainy,random_state=306,test_size=0.20)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_test))
```

# Cross Validation score for logistic regression:

```python
pred_lr=lr.predict(x_test)
from sklearn.model_selection import cross_val_score
lss=accuracy_score(y_test,pred_lr)
for j in range(2,30):
    lsscore=cross_val_score(lr,trainx,trainy,cv=j)
    lsc=lsscore.mean()
    print('At cv:-',j)
    print('Cross validation score is:-',lsc*100)
    print('accuracy_score is:-',lss*100)
    print('\n')
```

Since cross validation score accuracy score are almost same. We are heading with a good approach.

```python
lsscore_selected=cross_val_score(lr,trainx,trainy,cv=50).mean()
print('The cv score is :',lsscore_selected,'\nThe accuracy score ids :',lss)
```
